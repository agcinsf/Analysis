```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt


A = pd.read_csv('Pairwise_Data.csv', index_col=0)
A.rename(columns={'Which of the following topic areas is of most interest to you':'Categories'}, inplace=True)
```

```python
%matplotlib inline
import seaborn as sns
sns.set(style="whitegrid", color_codes=True)

#Group the data to see the total count of categories occuring
grouped = A.groupby('Categories').count().reset_index()

#plot the data to see visually what it looks like
plt.figure(figsize=(11, 8));
sns.set_context("notebook", font_scale=1.75, rc={"lines.linewidth": 2.5});
ax = sns.barplot(x='Categories', y='Timestamp', data=grouped.sort("Timestamp", ascending=False));
ax.set(ylabel='Total Observations');
```
